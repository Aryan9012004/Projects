```python
# Loading The required Library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading IPL Dataset

data = pd.read_csv('matches.csv')
```

```python
# First 5 rows
data.head()
```

```python
# Number of rows and columns
data.shape
```

```python
# Top 10 players with MOM(Man of the match) awards
data['player_of_match'].value_counts()[:10]
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(data['player_of_match'].value_counts()[:5].keys()),list(data['player_of_match'].value_counts()[:5]),color = "green")
plt.show()
```

```python
data['result'].value_counts()
```

```python
data['toss_winner'].value_counts()
```

```python
# Records where team won by batting runs
batting_first = data[data['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
# Histogram
plt.figure(figsize=(10,5))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of runs')
plt.xlabel('Runs')
plt.ylabel('Count')
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
plt.figure(figsize=(13,7))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),
        list(batting_first['winner'].value_counts()[0:5]))
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.pie(list(batting_first['winner'].value_counts()),
             labels=list(batting_first['winner'].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```

```python
batting_second = data[data['win_by_wickets']!=0]
```

```python
batting_second.head()
```

```python
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=20)
plt.show()
```

```python
batting_second['winner'].value_counts()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),
        batting_second['winner'].value_counts()[0:3],
       color=["purple","Blue","Red"])
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.pie(list(batting_second['winner'].value_counts()),
             labels=list(batting_second['winner'].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```
